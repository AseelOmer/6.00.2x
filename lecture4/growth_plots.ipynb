## ðŸ“Š Growth Order Visualization

In this notebook, we plotted four different types of growth functions:

- **Linear:** y = x
- **Quadratic:** y = xÂ²
- **Cubic:** y = xÂ³
- **Exponential:** y = 2Ë£

### ðŸ§  Observations:

- When plotted on the **same graph**, exponential growth dominates, making it difficult to see the linear and quadratic curves.
- By separating each function into its own plot using `plt.figure()`, we can clearly see:
  - Linear growth increases at a steady rate.
  - Quadratic and cubic growth accelerate faster, especially at higher x-values.
  - Exponential growth increases rapidly and quickly exceeds all others in scale.

### âœ… Conclusion:

Separating plots helps us **compare different scales** and better understand how each function behaves over time. This technique is crucial when working with data that grows at very different rates.


```python
import matplotlib.pyplot as plt

# Data
samples = list(range(30))
linear = [x for x in samples]
quadratic = [x**2 for x in samples]
cubic = [x**3 for x in samples]
exponential = [2**x for x in samples]

# Linear plot
plt.figure("Linear")
plt.plot(samples, linear)
plt.title("Linear Growth")

# Quadratic plot
plt.figure("Quadratic")
plt.plot(samples, quadratic)
plt.title("Quadratic Growth")

# Cubic plot
plt.figure("Cubic")
plt.plot(samples, cubic)
plt.title("Cubic Growth")

# Exponential plot
plt.figure("Exponential")
plt.plot(samples, exponential)
plt.title("Exponential Growth")

# Show all plots
plt.show()
```
